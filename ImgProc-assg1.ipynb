# 1- Documentation

### First, we import all required libraries, cv2 for opencv, matplotlib.pyplot to view the image, numpy required for matplotlib, we then remove the axis numbers
### Second, the showImg function is because openCV deal with images as BGR while matplot deals with RGB so we create a function to show the image automatically
### The HSV function is adjust the hue saturation and value, mainly for the saturation and value for the different filters, that happens by multiplying the hue and saturation by a certain coefficient, while for the value if to darken(value < 1) we multiply it directly, on the other hand, if to lighten(value >1) we use the formula from this [source](http://answers.opencv.org/question/193276/how-to-change-brightness-of-an-image-increase-or-decrease/) 
### The Lily function is for the lily filter, which adds a greenish tone to the image, with an increased brightness and a slight decrease in Saturation
### The lordKelvin function is for the lordkelvin filter, which adds an orange tone to the image, with an increased brightness and a slight decrease in Saturation
### The inkWell function is for the InkWell filter, which changes the image to grayscale, with a slight decrease in brightness
### The oneNine function is for the 1997 filter, which adds a very light pinkish tone to the image with a decrease in saturation to give the vintage look
### The popRocket function is for the PopRocket filter, which adds a pinkish tone to the image with a decrease in saturation
### The insta_like function is what the user can call, to call any of the previous functions according to the wanted to filter

# 2- Script

```python
%matplotlib inline
import cv2
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['xtick.major.size'] = 0
plt.rcParams['ytick.major.size'] = 0
plt.rcParams['xtick.labelsize'] = 0
plt.rcParams['ytick.labelsize'] = 0
```

```python
def showImg(img):
    temp = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    testimageplot = plt.imshow(temp)
    return temp

def HSV(img,h,s,v):
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    hsv[:,:,0] = hsv[:,:,0] * h
    hsv[:,:,1] = hsv[:,:,1] * s
    
    if(v<1):
        hsv[:,:,2] = hsv[:,:,2] * v
    elif(v>1):
        v2 = 2 - v
        hsv[:,:,2] = (hsv[:,:,2] * v2) + ((1-v2)*255)
    cimg = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)
    return cimg


def lily(image):
    img = image.copy()
    img[:,:,0] = img[:,:,0] *0.25 #+ 0
    img[:,:,1] = img[:,:,1] *1#+ 100
    img[:,:,2] = img[:,:,2] *0.85 #+ 50
    img = HSV(img,1,0.75,1.05)
    return img

def lordKelvin(image):
    img = image.copy()
    img[:,:,0] = img[:,:,0] * 0.25
    img[:,:,1] = img[:,:,1] * 0.75
    img[:,:,2] = img[:,:,2] 
    img = HSV(img,1,0.65,1.05)
    return img

def inkWell(image):
    img = image.copy()
    img = HSV(img,1,1,0.85)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    return img

def oneNine(image):
    img = image.copy()
    img[:,:,0] = img[:,:,0] * 1
    img[:,:,1] = img[:,:,1] * 0.8
    img[:,:,2] = img[:,:,2] * 1
    img = HSV(img,1,0.8,1.05)
    
    return img

def popRocket(image):
    img = image.copy()
    img[:,:,0] = img[:,:,0] * 0.45
    img[:,:,1] = img[:,:,1] * 0.25
    img[:,:,2] = img[:,:,2] * 1
    img = HSV(img,1,0.7,1)
    
    return img

def insta_like(image,instafilter):
    if(instafilter == 'Lily'):
        img = lily(image)
    elif(instafilter == 'Lord Kelvin'):
        img = lordKelvin(image)
    elif(instafilter == 'InkWell'):
        img = inkWell(image)
    elif(instafilter == '1977'):
        img = oneNine(image)
    elif(instafilter == 'Poprocket'):
        img = popRocket(image)
    
    return img
```
